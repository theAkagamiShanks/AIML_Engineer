# Prompt techniques

### **Prompting Techniques in AI (Prompt Engineering)**  

Prompting techniques are strategies used to **improve AI responses** by crafting better input prompts. These techniques help make **LLMs (like ChatGPT, GPT-4, etc.) more accurate, relevant, and useful**.

---

### **1️⃣ Zero-Shot Prompting**  
🟢 **What it is:** Asking a question **without giving any prior examples**.  
🟢 **When to use:** When you want AI to generate an answer based on its general knowledge.  

✅ **Example:**  
📝 **Prompt:** *"Explain machine learning in simple terms."*  
🤖 **Response:** *"Machine learning is a way for computers to learn from data and make decisions without being explicitly programmed."*  

🚀 **Pros:** Quick and efficient.  
⚠️ **Cons:** May lack accuracy or context.  

---

### **2️⃣ One-Shot Prompting**  
🟢 **What it is:** Providing **one example** before asking the AI to generate a response.  
🟢 **When to use:** When the model needs context but not extensive training.  

✅ **Example:**  
📝 **Prompt:**  
*"Translate the following to French: 'Good morning' → 'Bonjour'. Now translate: 'Thank you' →"*  
🤖 **Response:** *"Merci."*  

🚀 **Pros:** More accuracy than zero-shot.  
⚠️ **Cons:** Still limited learning from a single example.  

---

### **3️⃣ Few-Shot Prompting**  
🟢 **What it is:** Providing **multiple examples** before asking AI to generate an output.  
🟢 **When to use:** When the task requires patterns or style consistency.  

✅ **Example:**  
📝 **Prompt:**  
*"Translate the following to French:  
- 'Good morning' → 'Bonjour'  
- 'Thank you' → 'Merci'  
- 'How are you?' →"*  
🤖 **Response:** *"Comment ça va?"*  

🚀 **Pros:** Improves accuracy and relevance.  
⚠️ **Cons:** Can take longer and use more tokens.  

---

### **4️⃣ Chain-of-Thought (CoT) Prompting**  
🟢 **What it is:** Asking AI to **think step-by-step** to improve reasoning.  
🟢 **When to use:** For complex reasoning, math, logic, or problem-solving tasks.  

✅ **Example:**  
📝 **Prompt:**  
*"If a train leaves at 3:00 PM and travels 60 km per hour for 3 hours, how far does it go? Explain step by step."*  
🤖 **Response:**  
*"Step 1: The train travels at 60 km/hr.  
Step 2: It moves for 3 hours.  
Step 3: Distance = Speed × Time = 60 × 3 = 180 km.  
Final Answer: 180 km."*  

🚀 **Pros:** Improves reasoning and accuracy.  
⚠️ **Cons:** May generate lengthy responses.  

---

### **5️⃣ Role-Based Prompting**  
🟢 **What it is:** Asking AI to respond **from a specific perspective**.  
🟢 **When to use:** When you need a response in a particular style or expertise.  

✅ **Example:**  
📝 **Prompt:** *"You are a cybersecurity expert. Explain OWASP Top 10 risks in simple terms."*  
🤖 **Response:** *"As a cybersecurity expert, the OWASP Top 10 are the most critical web security risks... [detailed explanation follows]."*  

🚀 **Pros:** Makes responses more relevant and accurate for specific tasks.  
⚠️ **Cons:** Requires a well-defined role for effectiveness.  

---

### **6️⃣ Delimiters in Prompting (Formatting for Better Clarity)**  
🟢 **What it is:** Using **quotes, brackets, or sections** to structure a prompt for clarity.  
🟢 **When to use:** When dealing with **multi-step instructions or complex queries**.  

✅ **Example:**  
📝 **Prompt:**  
*"Summarize the following article:  
'''  
[Insert article here]  
''' "*  
🤖 **Response:** *[Summarized article]*  

🚀 **Pros:** Improves AI comprehension.  
⚠️ **Cons:** Needs careful formatting.  

---

### **7️⃣ Self-Consistency Prompting**  
🟢 **What it is:** Asking AI to generate multiple answers and pick the best one.  
🟢 **When to use:** When accuracy and reliability matter.  

✅ **Example:**  
📝 **Prompt:** *"Generate three different explanations for machine learning, then pick the best one."*  
🤖 **Response:** *"Here are three explanations... The best one is..."*  

🚀 **Pros:** Reduces errors, improves quality.  
⚠️ **Cons:** More computationally expensive.  

---

### **Final Thoughts**  
🚀 **Choosing the right prompting technique depends on your goal.**  
- For **general questions** → Use **Zero-Shot or One-Shot Prompting**.  
- For **pattern-based tasks** → Use **Few-Shot Prompting**.  
- For **complex reasoning** → Use **Chain-of-Thought Prompting**.  
- For **role-specific answers** → Use **Role-Based Prompting**.
