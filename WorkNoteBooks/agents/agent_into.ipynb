## Setting up agents

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import Tool
from langgraph.prebuilt import create_react_agent
```

```python
llm = ChatOpenAI(
    openai_api_base="http://localhost:1234/v1",
    openai_api_key="lm-studio",
    model="qwen2.5-0.5b-instruct",
    temperature=0
)
```

## Creating the Schema Structure

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel

class SubtractInput(BaseModel):
    a: int
    b: int

class AdditionInput(BaseModel):
    a: int
    b: int


def subtract(a: int, b: int) -> int:
    return abs(a - b)

def addition(a:int,b:int)->int:
    return sum([a,b])

subtraction_tool = StructuredTool.from_function(
    name="SubtractionTool",
    func=subtract,
    description="Subtract two numbers",
    args_schema=SubtractInput
)


addition_tool = StructuredTool.from_function(
    name="AdditionTool",
    func=addition,
    description="Use this tool to Add/Sum two numbers",
    args_schema=AdditionInput
)
```

```python
tools = [subtraction_tool,addition_tool]
```

```python

agent = create_react_agent(llm, tools)
```

```python
result = agent.invoke({"messages":["user","what is the sum of 45 ,1?"]})
```
