# Practicing RAG

```python
import langchain
from langchain_community.document_loaders import PyMuPDFLoader
```

## Loading the PDF

```python
file_path = "dataset/rag/Law.pdf"
loader = PyMuPDFLoader(file_path)
```

```python
docs = loader.load()
```

```python
docs[0].page_content
```

## Chunking

### Trying Sliding Window by increasing the Chunk Size and Chunk overlap


```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

slider = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=50
)

slide_chunks = slider.split_documents(docs)
```

```python
slide_chunks[0].page_content
```

```python
slide_chunks[1].page_content
```

## Saving into Vector DB 

```python

from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
embeddings.embed_documents
```

```python
faiss_store = FAISS.from_documents(slide_chunks, embeddings)
```

```python
faiss_store.save_local("models/fiass/faiss_index")
print("FAISS index saved!")
```
