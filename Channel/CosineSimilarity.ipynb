## Cosine Works!!!

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

sentences = [
    "I want to book a flight to Delhi",
    "Help me reserve a plane ticket to Delhi",
    "What is the capital of France?"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)

cos_sim = cosine_similarity(X)

print(cos_sim)
```

## Cosine FAILS (Repetition)

```python
sentences = [
    "Error occurred",
    "Error occurred error occurred error occurred error occurred"
]

vectorizer = TfidfVectorizer(use_idf=False)
X = vectorizer.fit_transform(sentences)

cos_sim = cosine_similarity(X)

print(cos_sim)
```

### Dot Product Works here

```python
import numpy as np

vectorizer = TfidfVectorizer(use_idf=False,norm=None)
X = vectorizer.fit_transform(sentences)


dot = (X @ X.T).toarray()

print("\nCosine similarity:")
print(cosine_similarity(X))


print("\nDot product:")
print((X @ X.T).toarray())
```

## Cosine FAILS (opposite meaning, same words)

```python
sentences = [
    "This medicine is effective",
    "This medicine is not effective"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)

print(cosine_similarity(X))
```

## Cosine FAILS (keyword overload)

```python
sentences = [
    "apple banana car",
    "apple banana car train phone laptop"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)

print(cosine_similarity(X))
```

### Jaccards works better here

```python
set1 = set(sentences[0].split())
set2 = set(sentences[1].split())

jaccard = len(set1 & set2) / len(set1 | set2)
print(jaccard)
```
